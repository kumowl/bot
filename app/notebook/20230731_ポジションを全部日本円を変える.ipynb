### ポジションを全部日本円に変えるノートブック

成り行き注文で全てのポジションを日本円に変える．   
実行間隔が短すぎて処理ができないことがあるので，その場合は再度実行するかsleepを挟む．

```python
import sys
sys.path.append('../')
import os
```

```python
import json
import pandas as pd

#自作モジュール
from gmo_api import GmoApi_private
```

```python
# config.jsonの読み込み
with open("../config.json", 'r') as f:
    config = json.load(f)

apiKey    = config['GMO_API_KEY']
secretKey = config['GMO_API_SECRET']
symbols = config['symbols']

gmo_api_private = GmoApi_private(apiKey, secretKey)

BASE_DIR = '../data/historical_klines/'
```

```python
position_df = pd.DataFrame(gmo_api_private.get_balance()['data'])
position_df[['amount', 'available', 'conversionRate']] = position_df[['amount', 'available', 'conversionRate']].astype(float)
position_df
```

```python
for symbol in symbols:
    size = position_df.loc[position_df['symbol'] == symbol]['amount'].values[0]
    gmo_api_private.post_order(symbol = symbol, side = 'SELL', executionType = 'MARKET', size = size)
```
